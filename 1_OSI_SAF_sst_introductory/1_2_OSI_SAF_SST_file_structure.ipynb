### Full resolution Metop-B Sea Surface Temperature:

Let's start by exploring the structure of a **level 2** product. We will here explore Metop-B/AVHRR data. 

```python
directory = os.path.join(os.getcwd(), 'products', 'L2_AVHRR_Metop-B')
```

Explore the directory and open the netcdf file

```python
for filename in os.listdir(directory):
    if filename.endswith(".nc"):
        nc = xr.open_dataset(os.path.join(directory, filename),decode_timedelta=False)
        print(filename)
```

We can now browse the file structure using xarray.
The following content will list the contents of the loaded file.

```python
nc
```

Have you had a detailed look at the diferent variables and attributes from the xarray output ?  
You can explore now explore individual variables. As an example :  
* sea_surface_temperature
* quality_level


```python
nc.variables['sea_surface_temperature']
```

```python
nc.variables['quality_level']
```

```python
nc.close()
```

### Global Metop Sea Surface Temperature :

We will now explore a level 3 product as well based on Metop-B/AVHRR measurements: the Global Metop Sea Surface Temperature.<br>
Define the dataset we want to explore

```python
directory = os.path.join(os.getcwd(), 'products', 'L3_Global_Metop')
```

Explore the directory and open the netcdf file. If you followed the chapter demonstrating the <a href="./1_1c_OSI_SAF_SST_data_access_IFREMER_FTP.ipynb" target="_blank">access to Ifremer FTP</a> and the one demonstrating the chapter demonstrating the chapter demonstrating the <a href="./1_1c_OSI_SAF_SST_data_access_Eumetsat_data_store.ipynb" target="_blank">access through the EUMETSAT data store</a>, you downloaded twice the same data with different filename. We will then here explore both separately.<br>
Let's start with the file from the FTP (<a haref="https://www.ghrsst.org/">GHRSST</a> compliant filename):

```python
keyword = 'OSISAF-L3C_GHRSST'

for filename in os.listdir(directory):
    if filename.endswith(".nc") and keyword in filename:
        nc = xr.open_dataset(os.path.join(directory, filename),decode_timedelta=False)
        print(filename)
```

We can now browse the file structure using xarray.
The following content will list the contents of the loaded file.

```python
nc
```

```python
nc.close()
```

Let's now have a look at the data from the EUMETSAT data store:

```python
keyword = 'S-OSI_-FRA_-'

for filename in os.listdir(directory):
    if filename.endswith(".nc") and keyword in filename:
        nc = xr.open_dataset(os.path.join(directory, filename),decode_timedelta=False)
        print(filename)
```

```python
nc
```

```python
nc.close()
```

As you noticed, the two Level 3 file structure are identical (both are GHRSST compliant). The only difference is the filename. <br>
Moreover, you will have noticed that the **level 2** and the **level 3** data structure are similar too. This is the case for most of SST datasets. The GHRSST has been working efficiently to create data specification, so that all SST data would be comparable. OSI SAF is taking part in this work, and take care to follow the <a href="https://www.ghrsst.org/resources/ghrsst-data-specification-gds-tables/">GHRSST data specification</a>.
